```python
results = requests.get(url).json()
results
```

```python
venues = results['response']['venues']

# tranform venues into a dataframe
dataframe = pd.json_normalize(venues)
dataframe.head()
```

```python
# keep only columns that include venue name, and anything that is associated with location
filtered_columns = ['name', 'categories'] + [col for col in dataframe.columns if col.startswith('location.')] + ['id']
dataframe_filtered = dataframe.loc[:, filtered_columns]

# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']

# filter the category for each row
dataframe_filtered['categories'] = dataframe_filtered.apply(get_category_type, axis=1)

# clean column names by keeping only last term
dataframe_filtered.columns = [column.split('.')[-1] for column in dataframe_filtered.columns]

dataframe_filtered
```

### The list ratings will contain the ratings of the restaurants and it will be made through a loop
### NOTE: SINCE THERE IS NO RATING ATTRIBUTE, THE NUMBER OF LIKES WILL BE THE ATTRIBUTE USED TO GENERATE THE RATING

```python
ratings = []

for i in ids:
    venue_id = i
    url = 'https://api.foursquare.com/v2/venues/{}?client_id={}&client_secret={}&v={}'.format(venue_id, CLIENT_ID, CLIENT_SECRET, VERSION)
    result = requests.get(url).json()
    
    rate = result['response']['venue']['likes']['count']
    
    ratings.append(rate)
```

```python
ratings
```

```python
df_final = dataframe_filtered
df_final['rating'] = ratings
df_final.drop('id', axis = 1, inplace=True)
df_final.head()
```

```python
df_final['rating'].values
```

### Now we have the desired Dataframe with all the features, we can plot the data in Folium

```python
seville_map = folium.Map(location=[latitude, longitude], zoom_start=14)
for lat, lng, name, rating in zip(df_final['lat'], df_final['lng'], df_final['name'], df_final['rating']):
    label = '{}, {}'.format(name, rating)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=4,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#87cefa',
        fill_opacity=0.5,
        parse_html=False).add_to(seville_map)
seville_map
```

### Lets divide the different restaurants into 5 categories according to their rating

```python
group = []

for index, row in df_final.iterrows():
    
    rate = row['rating']
    
    if rate < 6:
        group.append(1)
        
    elif rate >= 6 and rate < 20:
        group.append(2)
        
    elif rate >= 20 and rate < 30:
        group.append(3)
        
    elif rate >= 30 and rate < 40:
        group.append(4)

    else:
        group.append(5)
        
```

```python
df_final['group'] = group
df_final.head()
```

```python
df_final
```

### Now we can visually distinguish between the restaurants in the map

```python
seville_finalmap = folium.Map(location=[latitude, longitude], zoom_start=14)
for lat, lng, name, rating, group in zip(df_final['lat'], df_final['lng'], df_final['name'], df_final['rating'], df_final['group']):
    label = '{}, {}'.format(name, rating)
    label = folium.Popup(label, parse_html=True)
    
    if group == 1:
        folium.CircleMarker(
            [lat, lng],
            radius=4,
            popup=label,
            color='red',
            fill=True,
            fill_color='#87cefa',
            fill_opacity=0.5,
            parse_html=False).add_to(seville_finalmap)
    
    elif group == 2:
        folium.CircleMarker(
            [lat, lng],
            radius=4,
            popup=label,
            color='orange',
            fill=True,
            fill_color='#87cefa',
            fill_opacity=0.5,
            parse_html=False).add_to(seville_finalmap)
    
    elif group == 3:
        folium.CircleMarker(
            [lat, lng],
            radius=4,
            popup=label,
            color='yellow',
            fill=True,
            fill_color='#87cefa',
            fill_opacity=0.5,
            parse_html=False).add_to(seville_finalmap)
    
    elif group == 4:
        folium.CircleMarker(
            [lat, lng],
            radius=4,
            popup=label,
            color='cyan',
            fill=True,
            fill_color='#87cefa',
            fill_opacity=0.5,
            parse_html=False).add_to(seville_finalmap)
    
    elif group == 5:
        folium.CircleMarker(
            [lat, lng],
            radius=4,
            popup=label,
            color='green',
            fill=True,
            fill_color='#87cefa',
            fill_opacity=0.5,
            parse_html=False).add_to(seville_finalmap)
    
seville_finalmap
```
